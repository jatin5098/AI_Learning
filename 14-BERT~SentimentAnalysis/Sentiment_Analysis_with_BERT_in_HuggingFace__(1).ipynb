```python
def preprocess_function(examples):
  return tokenizer(examples["text"],padding=True,truncation=True)
```

```python
tokenized_dataset = dataset.map(preprocess_function, batched=True)
```

```python
tokenized_dataset
```

```python
### converting huggingface dataset into TF dataset

tf_train_dataset = tokenized_dataset["train"].to_tf_dataset(
    columns=['input_ids', 'token_type_ids', 'attention_mask', 'label'],
    shuffle=True,
    batch_size=BATCH_SIZE)
```

```python
tf_val_dataset = tokenized_dataset["test"].to_tf_dataset(
    columns=['input_ids', 'token_type_ids', 'attention_mask', 'label'],
    shuffle=True,
    batch_size=BATCH_SIZE)
```

```python
def get_positions(dataset):
  return {'input_ids':dataset['input_ids'],
          'token_type_ids':dataset['token_type_ids'],
          'attention_mask':dataset['attention_mask'],},dataset['label']
```

```python
tf_train_dataset=tf_train_dataset.map(get_positions).prefetch(tf.data.AUTOTUNE)
tf_val_dataset=tf_val_dataset.map(get_positions).prefetch(tf.data.AUTOTUNE)
```

# Modeling

## Based on TFBertForSequenceClassification

```python
model=TFBertForSequenceClassification.from_pretrained("bert-base-uncased",num_labels=2)
model.summary()
```

```python
# model_id
```

# Training

```python
num_epochs = 3
batches_per_epoch = len(tokenized_dataset["train"]) // BATCH_SIZE
total_train_steps = int(batches_per_epoch * num_epochs)
```

```python
optimizer, schedule = create_optimizer(init_lr=2e-5,num_warmup_steps=0, num_train_steps=total_train_steps)
```

```python
model.compile(loss=tf.keras.losses.BinaryCrossentropy(),
    optimizer=optimizer,
    metrics=['accuracy'])
```
